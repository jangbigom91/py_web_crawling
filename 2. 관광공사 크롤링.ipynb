# 관광공사 크롤링 후 txt로 저장

```python
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import sys

query_txt = query_txt = input('크롤링할 키워드 입력 : ')
f_name = input('검색결과를 저장할 파일 경로와 이름 입력 : (ex. /Users/jjink/Desktop/python/crawling_result/visit_korea_result.txt)')
path = '/Users/jjink/Desktop/python/chrome_driver/chromedriver'

driver = webdriver.Chrome(path)
driver.get('https://korean.visitkorea.or.kr/main/main.do#home')
time.sleep(2)

driver.find_element_by_class_name('btn_search').click()
element = driver.find_element_by_id('inp_search')
element.send_keys(query_txt)
driver.find_element_by_link_text('검색').click()

time.sleep(1)

full_html = driver.page_source # 현재 웹 페이지의 모든 html소스를 full_html에 저장
soup = BeautifulSoup(full_html, 'html.parser') # full_html에 저장된 소스코드를 BeautifulSoup가 분석하도록 지정
content_list = soup.find('ul', class_ = 'list_thumType')

for i in content_list :
    print(i.text.strip()) # list_thumType 의 텍스트만 추출 // 공백제거
    print('\n')
    
orig_stdout = sys.stdout # 표준 출력방향 변경(기본이 모니터 // orig_stdout으로 지정)
f = open(f_name, 'a', encoding = 'UTF-8') # 파일 경로, 인코딩 설정
sys.stdout = f # 출력방향 흐름 : 표준 출력방향 -> orig_stdout -> f
time.sleep(1)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
content_list = soup.find('ul', class_ = 'list_thumType')

for i in content_list :
    print(i.text.strip())
    print('\n')
    
sys.stdout = orig_stdout # f -> 표준출력 방향
f.close()

print('요청작업 완료')
```
