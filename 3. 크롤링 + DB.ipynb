```python
from bs4 import BeautifulSoup
from selenium import webdriver

import pymysql
import pandas as pd
import math
import time
import sys
```

```python
class ShowItems():
    def __init__(self):
        self.items = []
        self.db = pymysql.connect(host='127.0.0.1',
                                  port=3306,
                                  user='root',
                                  password='1234',
                                  db='visit_korea',
                                  charset='utf8')
        self.cursor = self.db.cursor(pymysql.cursors.DictCursor)
    
    def insert(self):
        sql = "INSERT INTO seoul VALUES (%(contents)s, %(tags)s);"
        self.cursor.executemany(sql, self.items)
        self.db.commit()
```

```python
query_txt = query_txt = input('크롤링할 키워드 입력 : ')
cnt1 = int(input('크롤링 할 건수 입력: '))
page_cnt = math.ceil(cnt1 / 10)

ft_name = input('검색결과 저장할 txt 파일 경로와 이름 입력 : (ex. /Users/jjink/Desktop/python/crawling_result/visit_korea_result.txt)')
fc_name = input('검색결과 저장할 csv 파일 경로와 이름 입력  : (ex. /Users/jjink/Desktop/python/crawling_result/visit_korea_result.csv)')
fx_name = input('검색결과 저장할 xlsx 파일 경로와 이름 입력 : (ex. /Users/jjink/Desktop/python/crawling_result/visit_korea_result.xlsx)')
print('\n')
path = '/Users/jjink/Desktop/python/chrome_driver/chromedriver'

driver = webdriver.Chrome(path)
driver.get('https://korean.visitkorea.or.kr/main/main.do#home')
time.sleep(2)

driver.find_element_by_class_name('btn_search').click()
element = driver.find_element_by_id('inp_search')
element.send_keys(query_txt)
driver.find_element_by_link_text('검색').click()

cnt2 = 0

time.sleep(1)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
real_page_cnt = math.ceil(cnt1 / 10)

print('\n')
print('=' * 80)
print('요청하신 건 수 : %s' %cnt1)
print('총 %s페이지의 게시글 수집 시작' %real_page_cnt)
print('=' * 80)
print('\n')

no1 = 1
no2 = []
contents2 = []
tags2 = []
showItems = ShowItems()

for x in range(1,real_page_cnt + 1) :
    print('==========%s페이지 내용 수집 시작==========' %x)
    print('\n')
    time.sleep(1)
    
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    content_list = soup.find('ul', class_ = 'list_thumType')
    try :
        content_list.find('li', class_ = "banner_area").decompose()
    except :
        pass
    contents = content_list.find('div', class_ = 'tit').get_text()
    tags = content_list.find('p', class_ = 'tag_type').get_text()
    
    for i in content_list :
        no2.append(no1)
        print('번호', no1)
        
        contents = i.find('div','tit').get_text()
        contents2.append(contents)
        print('내용', contents.strip())
        
        try :
            tag = i.find('p', 'tag_type').get_text()
        except :
            tag = '\n'
            tags2.append(tag)
            print('태그', tag.strip())
            print('\n')
        else :
            tags2.append(tag)
            print('태그', tag.strip())
            print('\n')
            
        showItems.items.append({'contents' : contents.strip(), 'tags' : tag.strip()})
        
        if no1 == cnt1 :
            break
            
        no1 += 1
        
    time.sleep(1)
    
    x += 1
    
    if x == real_page_cnt + 1 :
        break
        
    if (x % 6) == 0 :
        driver.find_element_by_xpath('//*[@id="%s"]' %x).click()
    else :
        driver.find_element_by_link_text("""%s""" %x).click()

showItems.insert()

korea_trip = pd.DataFrame()
korea_trip['번호'] = no2
korea_trip['내용'] = contents2
korea_trip['태그'] = tags2

korea_trip.to_csv(fc_name, encoding = 'utf-8-sig')
print('csv 파일 저장 경로 : %s' %fc_name)

korea_trip.to_excel(fx_name)
print('xlsx 파일 저장 경로 : %s' %fx_name)

f = open(ft_name, 'a', encoding = 'UTF-8')
f.write(str(contents2))
f.write(str(tags2))
f.close()
print('txt 파일 저장 경로 : %s' %ft_name)

driver.close()
```
