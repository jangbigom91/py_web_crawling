## 네이버 웹툰 제목 가져오기

* 월요 웹툰 영역

```python
from bs4 import BeautifulSoup as bs
from pprint import pprint
import requests
```

* 웹 페이지 열고 소스코드를 읽어오는 작업

```python
html = requests.get('https://comic.naver.com/webtoon/weekday.nhn')
soup = bs(html.text, 'html.parser')
```

* 월요 웹툰 영역 추출하기

```python
data1 = soup.find('div', {'class':'col_inner'})
pprint(data1)
```

* 제목 포함 영역 추출하기

```python
data2 = data1.findAll('a', {'class':'title'}) # 월요 웹툰영역 data1에서 findAll()을 사용해서 제목 영역 추출
pprint(data2)
```

* for를 이용한 텍스트 추출

```python
# 기본적인 for문으로 텍스트 추출
title_list = []
for t in data2:
    title_list.append(t.text)
    
pprint(title_list)
```

```python
# 리스트에서의 for문으로 텍스트 추출2
title_list = [t.text for t in data2]
pprint(title_list)
```

* 모든 요일 웹툰 제목 가져오기
* [요일별 웹툰영역 findAll] - [해당영역 제목 findAll] - [for로 text 추출]

* 요일별 웹툰영역 추출하기

```python
data1_list = soup.findAll('div', {'class':'col_inner'})
pprint(data1_list)
```

```python
# 전체 웹툰 리스트
week_title_list = []

for data1 in data1_list:
    # 제목 포함영역 추출하기
    data2 = data1.findAll('a', {'class':'title'})
    pprint(data2)
    
    # 텍스트만 추출
    title_list = [t.text for t in data2]
    #pprint("\n 웹툰 : " + str(title_list))
    
    week_title_list.append(title_list)
    
pprint(week_title_list)
```
